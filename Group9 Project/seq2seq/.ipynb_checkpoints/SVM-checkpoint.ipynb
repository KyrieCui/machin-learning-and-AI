```python
import random
import jieba
import pandas as pd
```

```python
punctuation = ["，", "。", "：", "；", "？"]
question_path = 'question.txt'
answer_path = 'answer.txt'
input_text = []
```

```python
with open(question_path, 'r', encoding='utf-8')as f:
    input_text = f.read().split('\n')
    #read the question
   
```

```python
input_text
```

```python
 #分词
segs_1 = [jieba.lcut(con) for con in input_text]
print(segs_1)
```

```python
for sentence in segs_1:
    while ' 'in sentence:
        sentence.remove(' ')
    # remove the space in each sentence
```

```python
print(segs_1)
```

```python
#remove punctuation 
tokenized=[]
for sentence in segs_1:
    words=[]
    for word in sentence:
        if word not in punctuation:
            words.append(word)
        
    tokenized.append(words)
print(tokenized)
len(tokenized)
```

```python
for question in tokenized:
    print(question)
```

```python
def preprocess_text(content_lines,sentences,category):
    sentences.append(("".join(content_lines),category))
```

```python
#give data labels from 0 to 1
sentences=[]
for i in range(121):
    #print(tokenized[i])
    #print(i)
    preprocess_text(tokenized[i],sentences,i)
    
   
```

```python
print(sentences)
len(sentences)
```

```python
#Break up the data set and generate more reliable training set distribution to avoid uneven distribution of similar data
random.shuffle(sentences)
sentences
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vec = CountVectorizer(
    analyzer='word', # tokenise by character ngrams
    max_features=120,  # keep the most common 1000 ngrams
)
vec
#Feature extraction, we define the feature of word bag model for text extraction
```

```python
#split the data into training data and test data
from sklearn.model_selection import train_test_split
x, y = zip(*sentences)
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=120)
```

```python
#Transform training data into word bag model
vec.fit(x_train)
```

```python
from sklearn.svm import SVC
svm = SVC(kernel='linear')
svm.fit(vec.transform(x_train), y_train)
print(svm.score(vec.transform(x_test), y_test))
print(x_test)
print(y_test)
```
